```python
def is_valid_chromosome(chromosome):
    # Check if chromosome is "X" or "Y"
    if chromosome in ["X", "Y"]:
        return True
    try:
        # Attempt to convert chromosome to an integer
        int(chromosome)
        return True
    except ValueError:
        # If conversion fails, chromosome is not valid
        return False

BED = BED[BED['Chromosome'].apply(is_valid_chromosome)]
BED['Chromosome'] = 'chr' + BED['Chromosome'].astype(str)
BED.shape
```

```python
import pandas as pd
from cyvcf2 import VCF

def parse_csq_header(vcf):
    for header in vcf.header_iter():
        info = header.info(extra=True)
        if info.get('ID') == 'CSQ':
            description = info['Description']
            format_description = description.split('Format: ')[-1].strip('"')
            return format_description.split('|')
    return []

def vcf_to_df(vcf_file):
    vcf = VCF(vcf_file)
    csq_header = parse_csq_header(vcf)
    samples = vcf.samples
    
    field_indices = {field: csq_header.index(field) for field in ['Consequence', 'IMPACT', 'Existing_variation', 'SYMBOL', 'Feature_type', 'Feature', 'gnomADe_AF', 'CADD_PHRED', 'ClinVar_CLNSIG', 'ClinVar_CLNREVSTAT', 'ClinVar_CLNDN'] if field in csq_header}
    
    records = []

    for variant in vcf:
        record = {
            'CHROM': variant.CHROM,
            'POS': variant.POS,
            'ID': variant.ID,
            'REF': variant.REF,
            'ALT': ','.join(variant.ALT),
            'QUAL': variant.QUAL,
            'FILTER': variant.FILTER,
        }
        
        for field in field_indices.keys():
            record[field] = []

        csq_annotations = variant.INFO.get('CSQ', [])
        for annotation in csq_annotations.split(','):
            fields = annotation.split('|')
            for field, index in field_indices.items():
                value = fields[index] if index < len(fields) else ''
                record[field].append(value)
                
        for key in field_indices.keys():
            record[key] = ','.join(set(filter(None, record[key])))
        
        for sample in samples:
            genotype_info_str = variant.format('GT')[vcf.samples.index(sample)]
            # Check if genotype is present and not './.:.:.:.:.'
            is_genotype_present = 1 if not all(gt == './.' for gt in genotype_info_str) else 0
            record[f'{sample}_GT_Presence'] = is_genotype_present

        records.append(record)
    
    return pd.DataFrame(records)

vcf_path = '/PATH/CVD-SNPs_VEP_annotated.vcf'
vcf = vcf_to_df(vcf_path)
```

```python
vcf_df = vcf.copy()
```

```python
vcf_df['CADD_PHRED'] = pd.to_numeric(vcf_df['CADD_PHRED'], errors='coerce')
vcf_df = vcf_df[vcf_df['CADD_PHRED'] >= 10]
vcf_df['gnomADe_AF'] = pd.to_numeric(vcf_df['gnomADe_AF'], errors='coerce')
vcf_df = vcf_df[vcf_df['gnomADe_AF'] <= 0.001]
vcf_df = vcf_df[vcf_df['QUAL'] >= 50]
vcf_df = vcf_df.drop('QUAL', axis = 1)
vcf_df = vcf_df.drop('FILTER', axis = 1)
vcf_df.shape
```

```python
import re

def extract_id(column_name):
    match = re.search(r'DNA-BR2-(\d+)_', column_name)
    if match:
        return match.group(1)  # Return the ID portion of the match
    return column_name

vcf_df.columns = [extract_id(col) for col in vcf_df.columns]

print(vcf_df.columns)
```

```python
vcf_df['Position'] = vcf_df['CHROM'].astype(str) + ':' + vcf_df['POS'].astype(str)
vcf_df['Allele'] = vcf_df['REF'].astype(str) + '>' + vcf_df['ALT'].astype(str)

# Rename columns
vcf_df.rename(columns={
    'Existing_variation': 'SNV',
    'SYMBOL': 'Gene',
    'gnomADe_AF': 'Allele Frequency',
    'CADD_PHRED': 'CADD Phred'
}, inplace = True)

new_column_order = ['Position', 'Allele'] + [col for col in vcf_df.columns if col not in ['Position', 'Allele', 'CHROM', 'POS', 'REF', 'ALT']]
vcf_df = vcf_df[new_column_order]
```

```python
def filter_snv_names(snv_str):
    snv_list = snv_str.split(',')
    filtered_snv_list = [snv for snv in snv_list if snv.strip().startswith('rs')]
    return ', '.join(filtered_snv_list)

vcf_df['SNV'] = vcf_df['SNV'].apply(filter_snv_names)
vcf_df.reset_index(drop=True, inplace=True)
CIGT_df = vcf_df.copy()
```

```python
vcf_df = vcf_df[['SNV', 'Position', 'Allele', 'Feature', 'Consequence', 'Allele Frequency', 'CADD Phred', 'Gene', 'ClinVar_CLNSIG', 'ClinVar_CLNREVSTAT', 'ClinVar_CLNDN']]
```

```python
# Category mapping
category_mapping = {
    '3_prime_UTR_variant': "3'-UTR",
    '5_prime_UTR_variant': "5'-UTR",
    'NMD_transcript_variant': "NMD Transcript",
    'downstream_gene_variant': "Downstream Gene",
    'intron_variant': "Intron",
    'mature_miRNA_variant': "Mature miRNA",
    'missense_variant': "Missense",
    'non_coding_transcript_exon_variant': "NC Transcript Exon",
    'non_coding_transcript_variant': "NC Transcript",
    'splice_acceptor_variant': "Splice Acceptor",
    'splice_donor_5th_base_variant': "Splice Donor 5th Base",
    'splice_donor_region_variant': "Splice Donor Region",
    'splice_polypyrimidine_tract_variant': "Splice Polypyrimidine Tract",
    'splice_region_variant': "Splice Region",
    'start_lost': "Start Lost",
    'stop_gained': "Stop Gained",
    'stop_retained_variant': "Stop Retained",
    'synonymous_variant': "Synonymous",
    'upstream_gene_variant': "Upstream Gene"
}

def clean_consequence(consequence):
    unique_consequences = sorted(set(
        category_mapping.get(part, part) 
        for part in consequence.replace('&', ',').split(',')
    ))
    return ', '.join(unique_consequences)

vcf_df['Consequence'] = vcf_df['Consequence'].apply(clean_consequence)

def add_space_after_commas(feature):
    return '|'.join([part.replace(',', ', ') for part in feature.split('|')])

vcf_df['Feature'] = vcf_df['Feature'].apply(add_space_after_commas)
vcf_df
```
